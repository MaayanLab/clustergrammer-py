# Export downsampling data 
The downsampling function will now export ds_data, downsampling data, which will enable users to track which original data points were assigned to downsampled clusters/data-points.

```python
from clustergrammer import Network
from clustergrammer_widget import clustergrammer_widget
```

```python
net = Network()
```

```python
net.load_file('txt/rc_two_cats.txt')
```

```python
df = net.export_df()
```

```python
net.load_df(df)
```

```python
ds_data = net.downsample(ds_type='kmeans', axis='row', num_samples=10)
```

```python
net.make_clust()
```

```python
clustergrammer_widget(network=net.widget())
```

# Make duplicate rows

```python
df = net.export_df()
```

```python
rows = df.index.tolist()
cols = df.columns.tolist()
```

```python
rows[0]
```

```python
new_rows = []
i = 0
for inst_row in rows:
#     inst_tuple = ('cluster', 'category '+ str(i))
    inst_tuple= 'cluster'
    i = i + 1
    new_rows.append(inst_tuple)
    
new_rows[0:3]
```

```python
new_cols = []
i = 0
for inst_col in cols:
#     inst_tuple = ('cluster', 'category '+ str(i))
    inst_tuple= 'col'
    i = i + 1
    new_cols.append(inst_tuple)
    
new_cols[0:3]
```

```python
df.columns = new_cols
```

```python
df.index = new_rows
```

```python
net.load_df(df)
```

```python
net.make_clust()
```
